## 2-D particle in a box (Lecture 3)

We saw in lecture 3 that the solutions for the 1-dimensional particle-in-a-box are

$$
\psi_{n}(x) = \sqrt{\frac{2}{L}}\text{sin}\left(\frac{n \pi x}{L} \right),
$$

where $n$ is the quantum number and $L$ is the box length.

Now consider a 2-dimensional box, with lengths $L_{1}$ and $L_{2}$. The Hamiltonian operator for a particle of mass $m$ is

$$
H = -\frac{\hbar^{2}}{2m} \frac{\partial^{2}}{\partial x^{2}} -\frac{\hbar^{2}}{2m} \frac{\partial^{2}}{\partial y^{2}}
$$

This is a separable Hamiltonian, of the form

$$
H = H_{x} + H_{y}
$$

so we know that the eigenfunctions of the Hamiltonian $H$ will just be the product of the eigenfunctions of the two independent "sub-Hamiltonians", $H_{x}$ and $H_{y}$. In other words, the eigenfunctions of the Hamiltonian $H$ are

$$
\psi_{nm}(x,y) = \sqrt{\frac{2}{L_{1}}}\sqrt{\frac{2}{L_{2}}} \text{sin} \left(\frac{n \pi x}{L_{1}} \right)  \text{sin} \left(\frac{m \pi y}{L_{2}} \right)
$$

where $n$ and $m$ are positive-integer quantum numbers greater than 1.

###Â What do these functions look like?

Consider the following code, which calculates and displays contour plots of $\psi_{nm}(x,y)$ for the case of a square box $L_{1} = L_{2} = L$.


```python
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

matplotlib.rcParams['xtick.direction'] = 'out'
matplotlib.rcParams['ytick.direction'] = 'out'
matplotlib.rcParams['contour.negative_linestyle'] = 'dashed'

#Â Set the box length.
Length = 1.0

#Â Set the step-length along each axis.
#
delta = 0.025

#Â Create arrays containing (x,y) positions.
#
x = np.arange(0.0, Length, delta)
y = np.arange(0.0, Length, delta)
X, Y = np.meshgrid(x, y)

#Â Calculate 2-dimensional particle-in-a-box wavefunctions for different (n,m).
#
Z11 = (2/Length) * np.sin(1 * np.pi * X/Length) * np.sin(1 * np.pi * Y/Length)
Z21 = (2/Length) * np.sin(2 * np.pi * X/Length) * np.sin(1 * np.pi * Y/Length)
Z12 = (2/Length) * np.sin(1 * np.pi * X/Length) * np.sin(2 * np.pi * Y/Length)
Z32 = (2/Length) * np.sin(3 * np.pi * X/Length) * np.sin(2 * np.pi * Y/Length)

#Â Plotting.
#
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2,sharex='col', sharey='row',figsize=(7,7))
CS = ax1.contour(X, Y, Z11, 6,colors='k')  
ax1.set_title('n=1, m=1')
ax1.set_xlim((0,1))
ax1.set_ylabel('y',fontsize=16)
CS = ax2.contour(X, Y, Z21,9,colors='k')
ax2.set_title('n=2, m=1')
ax2.set_xlim((0,1))
ax3.contour(X, Y, Z12,colors='k')
ax3.set_title('n=1, m=2')
ax3.set_xlabel('x',fontsize=16)
ax3.set_ylabel('y',fontsize=16)
ax4.contour(X, Y, Z32,colors='k')
ax4.set_title('n=3, m=2')
ax4.set_xlabel('x',fontsize=16)
plt.show()
```

In these plots, the solid lines denote positive regions of the wavefunction, and the dashed lines denote negative regions. As a result, there are obviously nodal planes which arise in the eigenfunctions; planes along which the probability of observing the particle is zero.

###Â What do the probability distributions look like?

The probability distributions are plotted below; these show peaks reflecting the eigenfunction plots above.

```python
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2,sharex='col', sharey='row',figsize=(7,7))
CS = ax1.contourf(X, Y, Z11**2)                  
ax1.set_title('n=1, m=1')
ax1.set_xlim((0,1))
ax1.set_ylabel('y',fontsize=16)
CS = ax2.contourf(X, Y, Z21**2)
ax2.set_title('n=2, m=1')
ax2.set_xlim((0,1))
ax3.contourf(X, Y, Z12**2)
ax3.set_title('n=1, m=2')
ax3.set_xlabel('x',fontsize=16)
ax3.set_ylabel('y',fontsize=16)
ax4.contourf(X, Y,Z32**2)
ax4.set_title('n=3, m=2')
ax4.set_xlabel('x',fontsize=16)
plt.show()
```
