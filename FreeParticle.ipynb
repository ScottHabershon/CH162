## Free particle quantum states (Lecture 4)

For a free particle of mass $m$ moving in 1-dimension, the Hamiltonian operator is

$$
\hat{H} = - \frac{\hbar^{2}}{2m} \frac{d^{2}}{dx^{2}}
$$

The corresponding free-particle solutions are of the form

$$
\psi(x) = e^{ikx}
$$

where $k$ is related to the momentum of the particle such that

$$
p = \hbar k.
$$

Note that, because we are dealing with a \textit{free} particle, there are no boundary conditions to impose quantization, so $k$ can take any value.

In this notebook, we will simply consider what these wavefunctions, and their probability distribution, look like.

```python
#Â Python setup
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Set up the wavefunction and set the k parameter.
#
k = 1.0
im = np.complex(0,1)
x = np.linspace(-10,10,100)
y = np.exp( im * k * x)
```

```python
#Â Plot real and imaginary parts of the wavefunction separately.
#
plt.plot(x,y.real,'r',lw=2,label='Real')
plt.plot(x,y.imag,'b',lw=2,label='Imaginary')
plt.xlabel('x',fontsize=16)
plt.ylabel(r'$\psi(x)$',fontsize=16)
plt.xticks(size=14)
plt.yticks(size=14)
plt.legend(fontsize=14)
plt.show()
```

From the graph above, we see that both real and imaginary parts of the wavefunction are oscillatory (sine and cosine) functions. 

But what does the probability distribution look like? Remember that the probability of obsrving a particle at $x$ is proportional to $\vert \psi(x) \vert^{2}$; this function is plotted for the free particle in the following graph.

```python
#Â Set up calculation of mod(psi)**2
#
plt.plot(x,(y.real**2+y.imag**2),'g')
plt.ylim([0,1.2])
plt.xlabel('x',fontsize=16)
plt.ylabel('P(x)',fontsize=16)
plt.xticks(size=14)
plt.yticks(size=14)
plt.show()
```

The conclusion is that the probability of finding the free particle at any point along the $x$-axis is just a constant!
